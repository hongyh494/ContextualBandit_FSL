# 单臂老虎机，在假定F(p)的回报相同时
1. 其中，$p$为触发老虎机赢钱的概率，$ \mu $应为触发赢钱的时候的平均回报，J为连续输钱J次后的J奖励，因此大样本下的单次摇一个老虎机的回报是
$$ F(p) = \mu * p + J * p_j \\ p_j = p\frac{(1-p)^j}{1-(1-p)^j} $$

2. 不同的J下，Casino在单臂下有个最小的p值s.t.每一把能贪到的money最多，但这个给出来以后顾客都知道


### mu = 0.4 

```python
import numpy as np

# F(p) = \mu * p + J * p_j

# numerator =  p * (1 - p)**(J)
# denominator = 1 - (1 - p)**(J)
# p_j = numerator / denominator
# asymptotic_mu_star = mu*p + J*p_j 
import matplotlib.pyplot as plt

# Given values for mu
mu = 1
mu = 0.4

# Generate a range of p values
p_values = np.linspace(0.01, 0.99, 100)

# Different values for J
J_values = [3, 4, 5, 6, 7]

# Plot the values for different J
for J in J_values:
    asymptotic_mu_star_values = []
    for p in p_values:
        numerator = p * (1 - p)**J
        denominator = 1 - (1 - p)**J
        p_j = numerator / denominator
        asymptotic_mu_star = mu * p + J * p_j
        asymptotic_mu_star_values.append(asymptotic_mu_star)
    plt.plot(p_values, asymptotic_mu_star_values, label=f'J={J}')

plt.xlabel('p')
plt.ylabel('Asymptotic Mu Star')
plt.title('Asymptotic Mu Star vs p for different J')
plt.legend()
plt.grid(True)
plt.show()
```

## 探究J变动的时候未来奖励的变化
